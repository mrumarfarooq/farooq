```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("mtcars.csv")
df.head()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='wt',y='mpg',data=df,marker='o',color='indigo')
plt.xlabel('wt')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.histplot(x='model',y='mpg',data=df,color='indigo')
plt.xlabel('model')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='cyl',y='mpg',data=df,color='indigo')
plt.xlabel('cyl')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.boxplot(x='am',y='mpg',data=df,color='indigo')
plt.xlabel('am')
plt.ylabel('mpg')
plt.show()
```
